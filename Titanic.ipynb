# **Categorical** **Values**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming titanic_data is already loaded with the Titanic dataset

# 1. Survival Count Plot
survival_counts = titanic_data['Survived'].value_counts()

# Plotting as a pie chart
plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title('Survival Proportion')
plt.show()

# 2. Survival Rate by Pclass
plt.figure(figsize=(6, 4))
sns.barplot(x='Pclass', y='Survived', data=titanic_data, ci=None)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# 3. Survival Rate by Sex
plt.figure(figsize=(6, 4))
sns.barplot(x='Sex', y='Survived', data=titanic_data, ci=None)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

```

**Insights**

The visualizations from the Titanic dataset reveal that **38.6%** of passengers **survived**, while **61.4% did not**. **Females** had a significantly higher **survival** rate than males. Most passengers were between **20 and 40 years old,** with children (0-17) and young adults (18-29) showing higher survival rates compared to middle-aged (30-49) and older passengers (50+). Additionally, **1st** class passengers had the highest **survival** rate, followed by those in 2nd class, with 3rd class passengers having the lowest survival rate.

# **Numerical** **Values**

**Age Distribution by Passenger** **Class**

```python

# 4. Age Distribution
plt.figure(figsize=(6, 4))
sns.histplot(titanic_data['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()



# 6. Survival Rate by Age Group
age_bins = [0, 18, 30, 50, 100]
age_labels = ['0-17', '18-29', '30-49', '50+']
titanic_data['AgeGroup'] = pd.cut(titanic_data['Age'], bins=age_bins, labels=age_labels)

plt.figure(figsize=(6, 4))
sns.barplot(x='AgeGroup', y='Survived', data=titanic_data, ci=None)
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()




# 9. Box Plots (e.g., Age and Pclass)
plt.figure(figsize=(6, 4))
sns.boxplot(x='Pclass', y='Age', data=titanic_data)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

**Observation**:

Most passengers were from the young age group 20-40 years.
The box plot shows the age distribution across different passenger classes (1st, 2nd, and 3rd class).
Insight: Passengers in the 1st class tend to be older on average compared to those in the 3rd class. The median age decreases from 1st to 3rd class, suggesting that younger passengers were more likely to be in lower classes.

**Survival** **Rate** **by** **Family** **Size**

```python
# . Survival Rate by Family Size (Siblings/Spouses Aboard)
plt.figure(figsize=(6, 4))
sns.barplot(x='Siblings/Spouses Aboard', y='Survived', data=titanic_data, ci=None)
plt.title('Survival Rate by Siblings/Spouses Aboard')
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Survival Rate')
plt.show()

# . Survival Rate by Family Size (Parents/Children Aboard)
plt.figure(figsize=(6, 4))
sns.barplot(x='Parents/Children Aboard', y='Survived', data=titanic_data, ci=None)
plt.title('Survival Rate by Parents/Children Aboard')
plt.xlabel('Number of Parents/Children Aboard')
plt.ylabel('Survival Rate')
plt.show()
```

Survival Rate by Siblings/Spouses Aboard

Insight: Passengers with 1 or 2 siblings/spouses aboard had higher survival rates compared to those with none or more than 2.

Survival Rate by Parents/Children Aboard

Insight: Passengers with 1 or 2 parents/children aboard had a higher survival rate. Those with no parents/children or more than 3 had lower survival rates.

**Fare** **Distribution**

```python
# . Fare Distribution
plt.figure(figsize=(6, 4))
sns.histplot(titanic_data['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```


**Observation**:

 The histogram shows a skewed distribution of fares with most passengers paying lower fares.
Insight: The majority of passengers paid fares below 100, with a few outliers paying much higher fares. This indicates a large disparity in ticket prices.

**Correlation** **Heatmap**



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming titanic_data is already loaded with the Titanic dataset


# 9. Correlation Heatmap
plt.figure(figsize=(8, 6))
# Select only numeric columns for correlation calculation
numeric_data = titanic_data.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()

```


Observation: The heatmap shows the correlation between various features (e.g., survived, Pclass, Age, SibSp, Parch, Fare).
Insight: There is a negative correlation between survival and Pclass, indicating that higher class passengers had a higher survival rate. Fare also shows a positive correlation with survival, suggesting that those who paid more were more likely to survive.

**Fare Distribution by Survival Status**

```python
import seaborn as sns
import matplotlib.pyplot as plt



# Separate data for survivors and non-survivors
survived = titanic_data[titanic_data['Survived'] == 1]
not_survived = titanic_data[titanic_data['Survived'] == 0]

# Plotting fare distribution for survivors and non-survivors
plt.figure(figsize=(10, 6))

sns.histplot(survived['Fare'], kde=True, color='skyblue', label='Survived', bins=30)
sns.histplot(not_survived['Fare'], kde=True, color='salmon', label='Not Survived', bins=30)

plt.title('Fare Distribution by Survival Status')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.legend()
plt.show()
```


Observation: The histogram shows fare distribution differentiated by survival status.
Insight: Survivors generally paid higher fares compared to those who did not survive, reinforcing the earlier observation about fare's positive correlation with survival
